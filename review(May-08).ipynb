이상치: 데이터의 전체적인 trend를 벗어나는 값을 이상치라고 판단한다.
> - 수량: -1
> - 상품번호: 999999 

통계적인 trend를 벗어나는 값을 판단하는 기준:
- 표준편차의 3배를 이상치라고 하자.
- 사분위 범위의 1.5배를 넘는 수를 이상치라고 하자.

- 컴퓨터는 수학자나 통계학자들이 만들어놓은 공식만 이용해서 연산을 한다.
그 공식들에는 이상치 데이터가 없다는 전제가 깔려있다. 
그래서 이상치를 적절히 처리하지 않고 계산하면 의미 있는 결과가 나오지 않는다. 
따라서 이상치 처리하고 통계적 데이터 분석을 해야 한다.

집단 : 데이터
- A : 2, 1, 3, 4, 5
> mean : 3
- B : 2, 1, 3, 4, 1000
> mean: 약 200, 1000이 이상치

```python
## 수량 이상치 확인 : filtering 사용

# 수량이 -1임에도 상품구매금액이 있다.
# 1) 잘못 입력
# 2) 오류
# 3) 환불 (단, 상품구매금액에도 음의값이 찍혀야 한다.)
# 가장 정확한 것은 이 data를 관리하는 실무자에게 물어보는 것이다.
# 이게 정말 이상치냐, 아니냐를 물어보는 것이 가장 정확하다.

cond1 = (df1["수량"] < 0)
df1.loc[cond1]
```

```python
## 수량 이상치 처리: -1이 아닌 data만 filtering하고 다시 describe 찍어본다.

# 수량 min: 1 (수량 이상치가 제거됬음을 확인)

df2 = df1.loc[~cond1]
df2.describe()
```

### 이상치

- 확인 방법
> - df.describe의 col별 min, max 값 확인 (이상치 조건 찾기)
> - cond1 = (df["col1"] 이상치 조건)
> - df.loc[cond1]


처리 방법(col 단위)
- 제거
> - cond = (df["col1"] 이상치 조건(연산자, 값))
> - df2 = df.loc[~cond1]

- 다른 값으로 대체
> - cond1 = (df["col1"] 이상치 조건)
> - df["col1(clean)"] = df["col1"] 
> - df.loc[cond1, "col1(clean)"] = 대체값
> - df.loc[cond1] 


- missing value로 대체
> - cond1 = (df["col1"] 이상치 조건)
> - df1.loc[cond1, "col1"] = np.nan

```python
cond1 = (df1["수량"] < 0)
df1.loc[cond1, "수량(clean)"] = 1
```

```python
df1.info()
```

```python
# 상품번호99999가 이상치 filtering
import numpy as np
cond1 = (df1["상품번호"] == 999999)
df1.loc[cond1, "상품번호"] = np.nan
cond2 = (df1["상품번호"].isnull())
df1.loc[cond2]
```

### 4-1 (Missing Value)

결측치 확인 -> 결측치 처리(dropna or new col) -> 처리된 df 사용 & 분석

- 확인 방법
> - df.info() : Non-Null Count는 col별 null이 아닌 val의 개수
> - df.isnull() : df의 val별 null 여부 확인 (if val is null, returns True)
>> df.loc[df["col1"].isnull()] 특정 col에 결측값이 있는 sample만 filtering
>  df.isnull().sum() : df의 col별 null-count 확인
>> [df.sum()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html)

- 처리 방법
> - df.dropna() : any row with null val을 drop (axis=0으로)
>> 데이터 loss가 너무 크다는 단점 존재
> - df.fillna() : df의 모든 null val을 대체할 값 지정
>> - df["col1"].fillna() : ser의 모든 null val을 대체할 값 지정
>> - df["col1(clean)"] = df["col1"].fillna(대체값)
>> - df["col1"].describe()의 mean 값을 긁어서 df["col1"].fillna(여기에 넣는다.)
>> - df2 = df.drop(columns=["col1"]): col1은 df2에 필요 없으므로 drop



- Data Cleansing : 데이터를 깔끔하게 다듬는 작업
> Data Cleaning의 첫 단계가 결측값 처리이다.
- 머신러닝 기법, 통계적 가설검정 기법을 원활하게 사용하기 위해서는 결측치를 반드시 처리해야 한다. 
- 결측값 종류: - NaN (Not a Number) / None / NA

```python
df1 = pd.read_csv("01_Contract_data.csv")
```

```python
# Non-Null Count 
df1.info()
```

```python
# True denotes the val is null.
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
# Credit_Rank의 val이 null인 df를 filtering
cond1 = (df1["Credit_Rank"].isnull())
df1.loc[cond1]
```

```python
## 방법1 : dropna
# 데이터 loss가 너무 크다. (51301 -> 40647)
df2 = df1.dropna()
df2
```

```python
## 방법2: fillna
cr_mean = df1["Credit_Rank"].mean()
df1["Credit_Rank(clean)"] = df1["Credit_Rank"].fillna(cr_mean)
df2 = df1.drop(columns=["Credit_Rank"])
df2["Credit_Rank(clean)"].isnull().sum()
```

ch3 핵심: indexing, sorting, filtering

```python
df1 = pd.read_csv("02_VIP_member_Data.csv")
```

```python
## 1. VIP이고 모바일 알람 수신하는 회원 filtering
cond1 = (df1["VIP회원"] == 'VIP')
cond2 = (df1["모바일알람여부"] == '수신')
df2 = df1.loc[cond1 & cond2]
#df2.to_excel("11.xlsx")
```

```python
## 2. VIP이고 총구매금액 상위 10명 filtering
df1.loc[cond1].sort_values(by="총구매금액", ascending=False).head(10)
```

```python
## 3. VIP이고 입력 받은 지역 filtering
address = "수지구"#input("Address : ")
cond3 = (df1["Address"] == address)
df1.loc[cond1 & cond3]
```

```python
## 4. VIP이고 입력 받은 연령대 filtering 하고 총구매금액이 높은 고객 순으로 sorting해서 저장
age_min = 35#int(input("Age min : "))
age_max = 45#int(input("Age max : "))

cond4 = (df1["Age"] >= age_min)
cond5 = (df1["Age"] <= age_max)

df1.loc[cond4 & cond5].sort_values(by="총구매금액", ascending=False)
```

```python
# 5. 총구매수량이 100건 이상이고, 총구매금액 상위 50명 filtering
cond6 = (df1["총구매수량"] >= 100)
df1.loc[cond6].sort_values(by="총구매금액", ascending=False).head(50)
```

```python
## 6a. Address항목에서 '수지구'에 사는 고객들의 VIP와 Member 수를 확인해 보세요!
cond1 = (df1["Address"] == "수지구")
df1.loc[cond1]["VIP회원"].value_counts()
```

```python
## 6b. Gender항목에 '남성' 고객들의 VIP와 Member수를 확인해 보세요!

cond1 = (df1["Gender"] == '남')
df1.loc[cond1]["VIP회원"].value_counts()
```

```python
## 6c. 배송서비스신청여부를 '신청' 또는 '미신청' 값을 입력하여,
# '총구매수량'이 높은 상위 100명의 명단을 
# 엑셀 파일로 추출하는 코드를 구성해보세요!

val = "신청"#input("배송서비스신청여부: ")
cond = (df1["배송서비스신청여부"] == val)
df2 = df1.loc[cond].sort_values(by="총구매수량", ascending=False).head(100)
#df2.to_excel("11.xlsx")
```
