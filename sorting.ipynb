## 숫자 데이터 Sorting 
- df.sort_values(by="col") : col 한개에 대해서 ascending sort(정렬)
> - df.sort_values(by="col", ascending=False) : descending sort (내림차순) 
- df.sort_values(by=["col1", "col2"]) : 선 col1 기준으로 ascending 정렬, 후 col2 기준으로 ascending 정렬
> - df.sort_values(by=["col1", "col2"], ascending=False) : 선 col1 기준으로 descending 정렬, 후 col2 기준으로 descending 정렬
- df.sort_values(by=["col1", "col2"], ascending=[False, True]) : 
- \ (back slash) 넣으면 줄바꿔도 코드 작동한다.
- 함수를 조합해서 우리가 원하는 형태로 데이터를 추출하는 방법
> - 선 indexing, 후 sorting
> - 선 sorting, 후 indexing

```python
# Amount Month: 월 비용
# 위 col 기준으로 낮은 것부터 높은 값까지 오름차순 (ascending) 정렬
df1.sort_values(by="Amount_Month")
```

```python
# descending sorted by Amount_Month
# 제품에 따라서 계약 금액이 결정된다.
# 그래서 같은 값이 어려개 나왔다. 215700
df1.sort_values(by="Amount_Month", ascending=False)
```

```python
#215700원 납부한 고객 중에, 나이가 높은 순서대로 정렬
# 두개 이상의 항목에 대한 정렬: by에 str이 아니라 list를 넣는다. 

# Amount_Month로 desceding sort를 한 다음에
# Age로 descending sort를 한다. 
# 두번 sort 하는데 list[0] 기준으로 먼저 sort 한다.

# 이렇게 하면 215700원으로 값이 중복될 때, Age를 기준으로 다시 정렬를 해주는 정렬.
df1.sort_values(by=["Amount_Month", "Age"], ascending=False)[["Amount_Month", "Age"]]
```

```python
# col1은 ascending, col2는 descending으로 sort 하는 방법
# Amount_Month를 descending sort 했을 때, 
# Amount_Month 값이 215700과 같이 중복될 경우,
# 중복되는 애들 간에 Overdue_Count를 기준으로 ascending sort
# ["Amount_Month", "Overdue_Count"]와 [False, True]의 index를 기준으로 적용하는 원리
df1.sort_values(by=["Amount_Month", "Overdue_count"], ascending=[False, True])
```

```python
# 여기에서 indexing(추출)까지 해보자
```

### indexing & sorting
- 선 sorting, 후 indexing
- 선 indexing, 후 sorting

```python
# 문제: Amount_Month값이 높으며, Age값이 높은 상위 100명의 고객명단을 구성하시오.
# df1.sort_values(by=["Amount_Month", "Age"], ascending=["False", "False"]).head(100)와 같음
df2 = df1.sort_values(by=["Amount_Month", "Age"], ascending=False).head(100)
```

```python
# 새로운 df2에 저장하면 Amount_Month값이 높으며, Age값이 높은 상위 100명의 고객명단이 df2에 저장된다.
```
