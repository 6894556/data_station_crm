## Sorting

### Numeric

```python
df.sort_values(by="Amount_Month")
```

```python
# Amount_Month를 기준으로 descending sort 결과: 215700원 납부한 고객들이 한명이 아님
df.sort_values(by="Amount_Month", ascending=False).head(20)[["Amount_Month", "Age"]]
```

```python
# 215700원 납부한 고객 여러 명을 나이 기준으로 descending sort
# 215700원 납부한 고객들을 descending sort
# by값이 lst라도 ascending=False로 해도 무관함
df.sort_values(by=["Amount_Month", "Age"], ascending=False)[["Amount_Month", "Age"]].head(21)
```

```python
# Amount_Month로 descending sort, Age로 ascending sort
# 215700원 납부한 고객들을 ascending sort
df.sort_values(by=["Amount_Month", "Age"], ascending=[False, True])[["Amount_Month", "Age"]].head(21)
```

## indexing

row 단위 
- df.head(n)
- df.tail(m) 
- df.iloc[r:s]

col 단위 
- df["col"] : ser 리턴
- df[["col1", "col2"]] : df 리턴
- df.loc[:, ["col1", "col2"]]

df 단위
- df.loc[n:m, ["col1", "col2"]]
- df[["col1", "col2"]].head(n)
- df.head(n)[["col1", "col2"]]
