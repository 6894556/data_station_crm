```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")
```

```python
df1 = pd.read_csv("dataset/01_Contract_Data.csv")
```

```python
# 1. Channel별 Amount_Month의 mean
#df1["Channel"].value_counts()
df1.pivot_table(index="Channel", values="Amount_Month", aggfunc="mean")
```

```python
# 2. Channel과 Gender별 Amount_Month의 mean
df1.pivot_table(index=["Channel", "Gender"], values="Amount_Month", aggfunc="mean")
```

```python
# 3. Gender와 Channel에 따른 Amount_Month의 sum
df1.pivot_table(index=["Gender", "Channel"], values="Amount_Month", aggfunc="sum")
```

```python
# 4. Gender와 Channel에 따른 Amount_Month와 Overdue_count의 sum
df1.pivot_table(index=["Gender", "Channel"], values=["Amount_Month", "Overdue_count"], aggfunc="sum")
```

```python
# 5. Gender와 Channel에 따른 Amount_Month와 Overdue_count의 sum, min, max
df1.pivot_table(index=["Gender", "Channel"], 
                values=["Amount_Month", "Overdue_count"],
               aggfunc=["sum", "min", "max"])
```

```python
# 6. Sales_Type에 따른 Datetime별 Amount_Month의 sum
df1.pivot_table(index=["Sales_Type"], columns="Datetime", values="Amount_Month", aggfunc="sum")
```

```python
df1["Datetime(dt)"] = pd.to_datetime(df1["Datetime"], format="%Y-%m-%d")
df1["Year"] = df1["Datetime(dt)"].dt.year
df1["Month"] = df1["Datetime(dt)"].dt.month
```

```python
# 7. Channel에 따른 월별  Amount_Month의 sum
df1.pivot_table(index="Channel", columns="Month", 
                values="Amount_Month",
               aggfunc="sum")
```

```python
# 8. Sales_Type에 따른 월별 Amount_Month의 mean
df1.pivot_table(index="Sales_Type",
                columns="Month", 
               values="Amount_Month",
               aggfunc="sum")
```

```python
# 9. Sales_Type에 따른 월별 Amount_Month의 sum (월별 매출의 추이)
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")
df1.pivot_table(index="Sales_Type", columns="Month", values="Amount_Month", aggfunc="sum")#.transpose().plot(subplots=True)
```

```python
# 10. Sales_Type에 따른 연도별, 월별 Amount_Month의 sum
df1.pivot_table(index="Sales_Type", columns=["Year", "Month"], values="Amount_Month", aggfunc="sum")
```

```python
# 11. Bank에 따른 월별 Amount_Month의 sum
#df1["Bank"].value_counts()
df1.pivot_table(index="Bank", columns="Month", values="Amount_Month", aggfunc="sum")
```

```python
# 12. Bank에 따른 월별 Amount_Month의 sum과 그들의 합 (margins=True)
df1.pivot_table(index="Bank", columns="Month", values="Amount_Month", aggfunc="sum", margins=True)
```

```python
# 13. Product_Type별 Amount_Month의 sum과 전체 합 (제품별합, 전체합)
df1.pivot_table(index="Product_Type", values="Amount_Month", aggfunc="sum", margins=True)
```

```python
# 14. Product_Type에 따른 연도별 Amount_Month의 sum (제품별합, 전체합) 
df1.pivot_table(index="Product_Type", columns="Year",
                values="Amount_Month", aggfunc="sum")#.to_excel("result2.xlsx")
```

```python
# 15. Product_Type에 따른 연도별 Amount_Month의 차이 (19년도 -> 20년도 증감액)
df3 = df1.pivot_table(index="Product_Type", columns="Year",
                values="Amount_Month", aggfunc="sum")

df3["증감액"] = df3[2020] - df3[2019]
df3
```

```python
# 16. 19년도 1-8월 데이터와 20년도 전체 데이터만 추출(filtering)하기
cond1 = (df1["Year"] == 2019)
cond2 = (df1["Month"] <= 8)
cond3 = (df1["Year"] == 2020)
df4 = df1.loc[(cond1 & cond2) | cond3]
```

### Big Picture
- pivot_tabel(index, columns, values, aggfunc)
> argument에 list를 넣을 수 있다.
- 정확한 제품별 매출 증감액 & 증감률 구하기
- 이번 시간: `pivot_table`로 데이터를 요약 & 정리
- 다음 시간: `melt`로 정리 & 요약된 데이터를 풀기
- 데이터를 뽑았을 때 정리된 형태로 뽑혔다면, 데이터를 풀어헤쳐야 한다. 
> 이 때, `melt`를 사용한다.

18. 1-8월 증감액 & 증감률
> - `증감률 = 증감액 / 전년도 + 이번년도`
> - 실제로 19년 대비 매출이 약 50%정도 대폭 감소했다. 


- index와 values, aggfunc에 list 형태 구조를 집어 넣을 수 있다.
- 동시에 2개의 범주형 항목으로 데이터를 나누어 2개의 숫자형 항목 데이터에 대한 통계량 계산
- aggfunc: default는 `numpy.mean`이다.
- `margins=True`: 총합 계산
> All이 index 마지막에 붙는다. 모든 은행 총합 값 (강력한 기능)
- Datetime: 영업일자, 계약일자 (csv 파일별 description 페이지가 필요함)
- erp(전사적 자원 관리): 데이터가 모여있는 공간
- 실무에서 많이 보는 항목: 전년 대비 매출 증감률, 고객 증감률
- pivot_table을 잘 사용하면, 다양하고 많은 데이터에서 데이터를 항목별로 나눠서, 특정 값에 대한 통계량을 확인하면서, 인사이트를 얻을 수 있다. 
- `df3[2020] - df3[2019]`: dt index가 아니라 col이름
> 제품별 매출 증감액 (음의 값이므로 줄었다.)
- 정확한 제품별 매출 증감액
> 19년도는 1-12월 데이터가 있지만 20년도는 1-8월 데이터만 있으므로 1-8월 증감만 계산해야 정확한 **증감액**과 **증감률**을 계산할 수 있다.
- 19년도, 20년도에 몇월부터 몇월까지의 데이터가 있는지 확인하는 방법:
> - 20년도:`cond1 = (df1["Year"] == 2020) df1.loc[cond1]["Month"].unique() # 20년도는 정확히 1-8월 데이터만 있음`
> - 19년도: `cond1 = (df1["Year"] == 2019) df1.loc[cond1]["Month"].unique() # 19년도는 정확히 1-12월 데이터가 있음`
- 19년도 1-8월 데이터와 20년도 전체 데이터만 추출(filtering)하기
> - `cond1 = df1["Year"] == 2019 cond2 = (df1["Month"] <= 8) cond3 = (df1["Year"] == 2020) df4 = df1.loc[(cond1 & cond2) | cond3]`
- 증감률이 0보다 작다는 것은 매출액이 줄었다는 것을 뜻한다.

```python
# 18. 1-8월 증감액 & 증감률
df5 = df4.pivot_table(index="Product_Type", columns="Year", values="Amount_Month", aggfunc="sum")
df5["증감액"] = df5[2020] - df5[2019]
df5["증감률"] = df5["증감액"] / (df5[2020] + df5[2019])
df5
```
