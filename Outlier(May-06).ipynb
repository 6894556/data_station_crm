통계적인 trend를 벗어나는 값의 기준 example
- e.g. 표준편차의 3배를 이상치라고 하자.
- 사분위 범위으  1.5배를 넘어가는 수를 이상치라고 하자.

- 컴퓨터는 수학자나 통계학자들이 만들어놓은 공식을 이용만해서 계산을 한다.
- 이상치 데이터를 포함해서 통계분석을 하면 안된다. 
- 이상치가 포함되서 계산하면 비정상적인 결과가 나온다.
- 따라서 이상치를 처리하고 데이터 분석을 이어서 해야 한다.

```python
## 수량 데이터 : describe에서 50이 이상치일까?라고 의심해볼 필요가 있다.
# 나중에 상품 구매 금액이랑 비교해보면 된다.
# 구매 금액이 크지 않은지
# 싼 물건 50개 구매했는지
# 데이터를 보면서 판단해야 한다.
df1["수량"]
```

```python
# 수량값이 0보다 작은 값들에 대해서 확인
cond1 = (df1["수량"] < 0)
df1.loc[cond1]
```

```python
# 수량이 -1임에도 상품구매금액이 있다.
# 1) 잘못 입력
# 2) 오류
# 3) 환불 (단, 상품구매금액에도 음의값이 찍혀야 한다.)
# 가장 정확한 것은 이 data를 관리하는 실무자에게 물어보는 것이다.
# 이게 정말 이상치냐, 아니냐를 물어보는 것이다.
```

```python
# 분석방법1 : 수량이 < 0 가 아닌 data만 갖고 분석한다.
df2 = df1.loc[~cond1]
```

```python
# 최소값이 1개부터 데이터가 찍혀있음을 볼수 있다.
# 이렇게 수량 이상치를 제거한 거다.
df2.describe()
```

```python
# 이상치가 포함된 행을 제거하면 그 행의 다른 정보들도 
# 다 제거가 되기 떄문에, 이상치를 missing value로 바꾸던
# 1로 바꾸던 값을 변환해서 분석하는 방법도 있다.

# 가장 정확한 것은 상품번호 확인해서 
# 그 상품의 금액과 수량을 구매금액과 매칭해보는 것이다.
df1.loc[cond1]
```

```python
# -1을 1로 바꿔보자.
# loc 함수를 사용해서
df1["수량(clean)"] = df1[""]
df1.loc[cond1, "수량(clean)"] = 1 # 1이라는 value를 assign
```

- 전략: 수량 -1을 1로 바꿔서 사용하자. 

```python
# cond1이 True인 data를 찾아서
# 수량(clean)이라는 새로운 col을 만들고
# 거기에 1이라는 값을 assign
# missing value
df1
```

```python
# cond1이 True인 행의 수량(clean) col에 1이라는 value를 assign

cond1 = (df1["수량"] < 0)

# 이렇게 해야 missing value가 없다.
df1["수량(clean)"] = df1["수량"]

df1.loc[cond1, "수량(clean)"] = 1
```

```python
# 기존 수량 col : -1
# New 수량(clean) col : 1

# -1만 1로 변환
# 개수도 같음
df1.describe()
```

```python
# 수량이 -1인 data만 filtering 해서
# 1로 변환 됬는지 안됬는지 확인해보기
cond1 = (df1["수량"] < 0)
df1.loc[cond1]
```

```python
# 이상치가 오면, 
# 범위를 줘서, loc 함수를 이용해 새 col을 생성해서
# 이상치를 대체값으로 변환할 수 있다.
```

이상치 처리 방법
- 방법1 : 이상치 제거하고 새로운 df 만들어서 분석
> 이상치가 있는 행을 제거하고 분석
- 방법2 : 기존 col의 이상치를 다른 값으로 대체한 새로운 col을 생성 후 분석
> 이상치를 다른 값으로 대치하고 분석

```python
# 상품번호99999가 이상치인지 확인해보기
cond2 = (df1["상품번호"] == 999999)
df1.loc[cond2]
# 상품 구매금액 col과 비교해보기: 다 다름
# 배송시작일 col : 주문 시간이 비슷함

# 따라서 잘못 찍힌 것으로 판단.

# 방법1: 이 데이터만 제외 후 분석
# 방법2: 상품번호를 missing value값을 assign
```

```python
import numpy as np
```

```python
# np : numeric python
# missing value 집어넣기 : np.nan
cond2 = (df1["상품번호"] == 999999)
# 상품 번호가 9 여섯개인 데이터의 상품번호를 missing value로 대체
df1.loc[cond2, "상품번호"] = np.nan 
```

```python
#df1.loc[cond2]
# 9 여섯개가 missing value로 대체됬는지 확인
cond3 = (df1["상품번호"].isnull())
df1.loc[cond3]
```

```python
# 기존 col에 다른 val을 대치하면 
# 원래 값을 다시 확인할 수가 없다.
# 따라서, 값을 대치해서 쓸 때는
# 새로운  col 만들어서 사용하는게 가장 좋다.
```
