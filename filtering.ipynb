```python
# boolean Series
# 왜 tuple로 안하면 error가 발생하는지 모르겠다.
# 왜 &가 아닌 and OPERATOR 사용하면 error가 발생하는지 모르겠다.
# df의 모든 idx에 대해서, boolean값 1개를 return하는 ser
# parenthes(소괄호)로 감싼다고 tuple이 되는 것도 아니다. ser이다.
(df["Address1"] == "경기도") & (df["Overdue_count"] >= 5)
```

```python
# Adress1이 경기도이면서, Overdue_count가 5회 이상인 데이터 filtering
# 꼭 loc indexer를 사용하지는 않아도 된다.
#df[(df["Address1"] == "경기도") & (df["Overdue_count"] >= 5)] : error 발생 안함
# conditions를 계속 받아 쓰려니 귀찮긴 하다.
cond1 = (df["Address1"] == "경기도")
cond2 = (df["Overdue_count"] >= 5)
df.loc[cond1 & cond2]
```

```python
# Age가 30세 미만이면서, Overdue_count 5회 이상인 데이터 filtering
cond1 = (df["Age"] < 30)
cond2 = (df["Overdue_count"] >= 5)
df.loc[cond1 & cond2]
```

```python
# Address1이 경기도이거나, 충청도인 고객 데이터 filtering
cond1 = (df["Address1"] == "경기도")
cond2 = (df["Address1"] == "충청도")
df1 = df.loc[cond1 | cond2]
df1
```

```python
df1["Address1"].value_counts()
```

```python
# 렌탈 유형의 고객 데이터만 filtering 후, 연체건수가 5회 이상인 고객 filtering
cond1 = (df["Sales_Type"] == "렌탈")
cond2 = (df["Overdue_count"] >= 5)
df.loc[cond1 & cond2].sort_values(by="Overdue_count", ascending=False).head(10)
```
