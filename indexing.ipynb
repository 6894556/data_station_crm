## indexing (원하는 데이터 추출하기)
- 행단위 추출(행(row, index)단위로 데이터 가져오기)
> - df.head() : 상위 5개 데이터 추출
> - df.tail(): 하위 5개 데이터 추출
> - df.iloc[] : 원하는 범위 데이터 행단위 추출
- 열단위 추출(col)
> - df["col"] : ser 형태로 열 단위 추출 (col 하나만 열단위로 indexing 때 사용)
> - df[["col"]] : df 형태로 열 단위 추출 (col 두개 이상 열단위로 indexing할 때 사용)
>> - df[["col1", "col2"]] : df 형태로 col1, col2 추출



실무에서 5만 건에 20개 열은 데이터 양이 많은 게 아니다. 
공장에서 data를 받아오면, 열 개수만 50개가 넘는 경우가 많다. 
즉, 실무자의 기준에서도 데이터 양이 많은 상황이 있다.
이런 상황에서 indexing을 사용해서 본인이 분석할 수 있는 data만
행 단위 indexing 또는 열 단위 indexing을 사용해서 추출해야 한다. 

<br>
내가 분석할 수 있는 데이터만 indexing으로 가져와서 분석하는 것이 가장 효과적이다. 
그래야 보기도 쉽고 다루기도 어렵지 않다. 
<br>

실무에서는 indexing해서 분석해야 한다.

```python
df1.iloc[:50] # df1.head(50)과 같은 df를 리턴
```

### 열단위 데이터 추출

```python
df1["Age"] # ser 형태(index-value)로 추출
```

```python
df1[["Age"]] # df 형태(index-column-value)로 추출
```

```python
df1.columns
```

```python
# 여러 열을 ser 형태로 추출하면 error가 발생하는 이유:
# Ser는 index-value 구조로 구성 되어 있음
# 따라서 Ser는 index-column-value 구조를 이해할 수 없음
# 그러므로 col 두 개 이상 indexing을 해야 하는 경우, 
# Ser가 아닌 df로 indexing을 해야 함. 
df1["Sales_Type", "Contract_Type", "Amount_Month"]
```

```python
df3 = df1[["Sales_Type", "Contract_Type", "Amount_Month"]]
```

```python
# 조합해서 사용하기 : 이런식으로 여러 함수를 조합해서(이어서) 사용할 수 있다.
df3.head(100)
```

```python
# 저 col 두개 상위 1천개만 새로운 변수에 선언하고 싶다: 열 두개 indexing하고 1천 개 추출
df4 = df1[["Product_Type", "Bank"]].head(1000)
df4
```

```python
# 순서를 반대로 해보기: 1천개 먼저 추출하고 열 두개만
df1.head(1000)[["Product_Type", "Bank"]]
```
