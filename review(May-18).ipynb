- crm 같은 데이터는 mean 보다는 sum을 더 많이 계산한다.
- pivot_table(aggfunc): 특정 통계량을 계산하는 기능
- pivot_table(margins): row별, col별 총합 계산하는 기능 (All)

### 구분하고 싶은 항목이 하나의 범주형 항목인 경우

int는 col의 개수
- index : categorical
> unique 카운트가 적은 것부터 기재해야 편함
- values: numeric col (aggfunc 사용해야 하므로)
- columns:
- index, values, aggfunc에 list 형태 구조를 넣을 수 있다
- Channel별 월 Amount_Month의 mean, sum



- index=1, values=1, aggfunc=1
> - Product_Type에 따른 Amount_Month의 sum과 매출 총액(margins)
- index=1, values=1, columns=1, aggfunc=1
> - Product_Type에 따른 연도별 Amount_Month의 sum
> - Sales_Type에 따른 dt.month별 Amount_Month의 mean 또는 sum (19년, 20년 n월 데이터가 합쳐진 sum)
> - Bank에 따른 월별 Amount_Month의 sum
- index=1, values1, columns=2, aggfunc=1
> - 19년, 20년 n월 데이터가 나눠진 sum (반드시 Year, Month 순으로 기재해야 함)

### 구분하고 싶은 항목이 두개 이상의 범주형 항목인 경우
- index=2, values=1, aggufunc=1
- index=2, values=2, aggfunc=1
> - Gender와 Channel에 따른 Amount_Month의 mean 또는 sum
> - Gender와 Channel에 따른 Amount_Month와 Overdue_count의 mean 또는 sum
- index=2, values=2, aggfunc=3
> - Gender와 Channel에 따른 Amount_Month와 Overdue_count의 sum, min, max


```python
df1.pivot_table(index="Channel", values="Amount_Month", aggfunc="sum")
```

```python
# unique 카운트가 적은 것부터 기재
df1.pivot_table(index=["Gender", "Channel"], values="Amount_Month", aggfunc="mean")
```

```python
df1.pivot_table(index=["Gender", "Channel"],
                values=["Amount_Month", "Overdue_count"],
               aggfunc=["sum", "min", "max"])
```

```python
df1["Datetime(dt)"] = pd.to_datetime(df1["Datetime"], format="%Y-%m-%d")
df1["Year"] = df1["Datetime(dt)"].dt.year
df1["Month"] = df1["Datetime(dt)"].dt.month

# 19년, 20년 n월 데이터가 합쳐진 sum
df1.pivot_table(index="Sales_Type", columns="Month", values="Amount_Month", aggfunc="sum")
```

```python
df1.pivot_table(index="Sales_Type", columns=["Year", "Month"], values="Amount_Month")
```

```python
df1.pivot_table(index="Bank",
                columns="Month",
                values="Amount_Month",
                aggfunc="sum", 
                margins=True)
```
