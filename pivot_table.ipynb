```python
import pandas as pd
```

```python
df1 = pd.read_csv("01_Contract_Data.csv")
```

pivot_table 사용법
- excel에도 pivot이라는 기능이 있다.
> 항목 1에 따라서, 특정 값을 계산해주는 기능
- 성별에 따라서 매출액 차이가 있는가?
- 결제은행에 따라서 연체건수에 차이가 있는가?
구분할 수 있는 범주형 항목에 따라서 숫자 데이터의 통계량을 계산하는 함수 

pivot_table 쓰기 전에 groupby, pivot 함수를 쓴다.
- 실무에서는 pivot_table 사용하면 대부분의 데이터를 항목별로 계산 or 정리를 해줄 수 있다.

```python
df1.info()
```

```python
# 유입 채널에 따라서 월 렌탈 비용에 차이가 있는가?
# 채널별 월 결제액의 평균
# crm 같은 데이터는 mean 보다는 sum을 더 많이 계산한다.
df1.pivot_table(index="Channel", values="Amount_Month", aggfunc="mean")
```

```python
# aggfunc: 특정 통계량을 계산하는 기능
df1.pivot_table(index="Channel", values="Amount_Month", aggfunc="sum")
```

```python
# pivot table 사용하면 다양한 분석이 가능해진다.
# 구분하고 싶은 항목이 두 개 이상의 범주 항목인 경우
# e.g. 채널과 성별에 따라서 월 결제액
df1.pivot_table(index=["Channel", "Gender"], values="Amount_Month", aggfunc="mean")
```

- index와 values, aggfunc에 list 형태 구조를 집어 넣을 수 있다.

```python
# 성별에 따른 채널별 월 렌탈비용의 합 
df1.pivot_table(index=["Gender", "Channel"], values="Amount_Month", aggfunc="sum")
```

```python
# overdue count : 연체건수
# 동시에 2개의 범주형 항목으로 데이터를 나누어 2개의 숫자형 항목 데이터에 대한 통계량 계산
df1.pivot_table(index=["Gender", "Channel"], values=["Amount_Month", "Overdue_count"])
```

```python
# 성별과
# 채널에 따른
# Amount_Month와 Overdue_count의 합, 최소값, 최대값을 한번에 계산 
df1.pivot_table(index=["Gender", "Channel"], values=["Amount_Month", "Overdue_count"], aggfunc=["sum", "min", "max"])
```

- columns: 

```python
# Datetime: 영업일자, 계약일자
# 일자별로 Sales_Type에 따른 Amount_Month의 합 (매출액)
# sales type에 따라서 data가 찍힌다: 577일 -> 너무 많다.
df1.pivot_table(index="Sales_Type", values="Amount_Month", columns="Datetime", aggfunc="sum")
```

```python
# 577 columns가 너무 많다. 
# 그러니 Datetime을 datetime 처리해서 
# 주차별, 월별, 연도별 지출액을 계산해보자
df1["Datetime(dt)"] = pd.to_datetime(df1["Datetime"], format="%Y-%m-%d")
```

```python
# Year, Month 열 만들고 연도별, 월별 지출액 계산
df1["Year"] = df1["Datetime(dt)"].dt.year
df1["Month"] = df1["Datetime(dt)"].dt.month
```

```python
# 월별로 판매 채널에 따른 월 요금의 평균: 
df1.pivot_table(index="Sales_Type", columns="Month", values="Amount_Month")
```

```python
# 월별 매출의 추이: 19년도 1월 데이터와 20년도 1월 데이터가 합쳐져서 계산이 된 것이다.
df1.pivot_table(index="Sales_Type", columns="Month", values="Amount_Month", aggfunc="sum")
```

```python
# 19년, 20년을 나눠서 월별 매출 합 계산
df1.pivot_table(index="Sales_Type", columns=["Year", "Month"], values="Amount_Month", aggfunc="sum")
```

```python
# erp라는 데이터가 모여있는 공간에서 데이터를 가져와서
# 이런 식으로 보고서를 써야하는 경우가 많이 있다. weekly report, monthly report
# python 코드로 pivot table로 logic을 만들어 놓으면 
# 매주 똑같은 작업을 반복해서 할 필요 없이
# 데이터 가져와서 실행만 시키면 
# 알아서 데이터가 처리 되고 파일로 저장까지 된다. -> 자동화
```

```python
# 결제 은행별 월별 결제 금액의 합이 어떻게 되는가?
df1.pivot_table(index="Bank", columns="Month", values="Amount_Month", aggfunc="sum")
```

- margins=True : 총합 계산 
> All이 index 마지막에 붙는다. 모든 은행 총합 값

강력한 기능

```python
# 결제 은행별 월별 결제 금액의 합의 총합
df1.pivot_table(index="Bank", values="Amount_Month", aggfunc="sum", margins=True)
```

```python
# Product_Type별 Amount_Month의 sum과 매출액의 총액 : 제품별합, 전체합
df1.pivot_table(index="Product_Type", values="Amount_Month", aggfunc="sum", margins=True)
```

```python
# 제품별합, 전체합을 연도별로 보겠다면: columns="Year" 
# + 연도별 총합도 계산된다.
df1.pivot_table(index="Product_Type", values="Amount_Month", aggfunc="sum", columns="Year")
```
