### 숫자데이터 확인하기
실무에서 확인하는 사항들
- 평균: 숫자데이터의 대표값(중앙값)을 확인 (오만건의 데이터를 대표할 수 있는 숫자 1개)
- 합 : 숫자데이터의 전체 합 확인 (e.g 매출액)

- 숫자 데이터가 굉장히 많은데, 이거를 일일이 다 확인해야 하는가?
> 다 확인하면 좋다. 하지만 수가 너무 많으면 일일이 확인하기 어렵다.
> 이 때, describe()를 사용해서 확인한다.
> 숫자 데이터 다 갖고와서 요약 통계량(평균, 표준편차, 최소값, 최대값, ... )을 확인해준다. 

- 합을 확인할 때는 df1.sum(axis=1)

```python
# 내 코드
import numpy as np
df1.select_dtypes(include=np.number)
```

## 문자 데이터 확인하기
- 항목 : 어떤 항목이 있는가?
- 빈도수 : 그 항목이 몇개씩 구성되어 있는가?

- 한 항목에 몇가지 unique 값이 있는지 detail하게 확인하려면 df1.value_counts()를 사용해야 한다
- 기본적인, 전반적인 정보만 확인하려면 df1.describe(include="object)를 사용한다.

```python
df1["Sales_Type"] # ser형태로 문자 데이터를 뽑는다 : object
```

```python
# Sales_Type 항목에 어떤 고유 value들이 있는지 보여준다
df1["Sales_Type"].unique() 
```

```python
# 어떤 은행을 사용했는가?
df1["Bank"].unique()
```

```python
# 렌탈은 몇건, 멤버십은 몇건 있는가? : 빈도수
df1["Sales_Type"].value_counts() # 문자 데이터 항목의 빈도수 계산
```

```python
# 각 은행별로 몇명이 사용했는가? : 빈도수
df1["Bank"].value_counts()
```

```python
# 숫자의 describe 같은 함수가 문자 데이터에는 없는가? 
# 문자 데이터만 가져와서 기본적인 요약 통계량 확인 : 문자 데이터의 요약 통계량 확인

# count: 개수
# unique: unique value의 개수
# top : unique value 중 가장(top) 빈도 수가 높은 object type의 value -> 렌탈이라는 값이 제일 많다. (최고 빈도수를 갖는 값)
# freq : 가장 빈도수가 많은 unique value의 frequency만 출력

df1.describe(include="object")
```
