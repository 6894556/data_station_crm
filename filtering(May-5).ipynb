```python
# 특정 조건에 부합하는 데이터만 가져오기
df1["Contract_Type"].value_counts()
```

### Contract_Type이 교체 계약인 고객 데이터만 filtering
- filtering은 row 단위 추출
- 몇 rows인지 확인 필요

```python
# 여러가지 계약 유형 중 : value_counts
# Contract_Type이 교체 계약인 고객만 필터링
```

```python
# boolean series : 모든 index에 대해서 Contract_Type의 value가 교체계약이면 True (index-value)
df1["Contract_Type"] == "교체계약"
```

```python
# = : assign operator
# == : boolean operator
condition1 = df1["Contract_Type"] == "교체계약"
condition1
```

```python
df1.loc[condition1]
```

### Amount_Month가 150000 이상인 고객 데이터만

```python
cond2 = df1["Amount_Month"] >= 150000
df1.loc[cond2]
```

### Overdue_count가 5회 이상인 고객 데이터만 filtering
- 연산자가 여러 개 들어가면 헷갈릴 수 있으니 소괄호로 묶는다.
- condition의 약자는 cond

```python
cond3 = (df1["Overdue_count"] >= 5)
df1.loc[cond3]
```

## 조건 두 개 이상 만족하는 데이터 가져오기
- df.loc[cond1 & cond2]
- df.loc[cond1 | cond2]
- 함수를 여러개 붙여서 써서 우리가 원하는 데이터를 빠르고 쉽게 가져올 수 있다. (이 작업을 연습해야 한다.)

### Address1이 경기도이면서 Overdue_count가 5회 이상인  고객 데이터 filtering
경기도에 살면서 연체건수가 5건 이상인 고객들만

```python
cond4 = (df1["Address1"] == "경기도")
df1.loc[cond4 & cond3]
```

### Age가 30 미만이면서 Overdue_count가 5회 이상인 고객 데이터 filtering

```python
cond5 = (df1["Age"] < 30)
df1.loc[cond5 & cond3]
```

### Address1이 경기도이거나 충청도인 고객 데이터만 filtering

```python
cond6 = (df1["Address1"] == "충청도")
df1.loc[cond4 | cond6]
```

### Sales_Type이 렌탈인 고객 데이터만 filtering 후, Overdue_count가 5회 이상인 고객 데이터만 filtering

```python
cond7 = (df1["Sales_Type"] == "렌탈")
df1.loc[cond7 & cond3].sort_values(by="Overdue_count", ascending=False).head(10)
```

### Sales_Type이 렌탈인 고객 들 중에서 Overdue_count 상위 10명 고객 데이터 filtering


```python

df1[cond7].sort_values(by="Overdue_count", ascending=True).head(10)
```
