### 여러 개 병합

```python
# 파일명이 비슷하다. 따라서:
for i in ['1', '2', '3', '4']:
    print("2016_health_checkup_"+ i +".csv")
```

```python
df_raw = pd.DataFrame()
for i in ['1', '2', '3', '4']:
    df_ = pd.read_csv("2016_health_checkup_"+ i +".csv")
    df_raw = pd.concat([df_raw, df_])

# loop 이해
# empty, 1
# empty1, 2
# empty12, 3
# empty123, 4
```
