```python
# 지역을 입력하면 그 지역 vip 고객 명단만 뽑아보자

value1 = input("Address 입력하시오: ") # string


cond1 = (df1["VIP회원"] == "VIP")
cond2 = (df1["Address"] == value1)

df1.loc[cond1 & cond2]
```

```python
# 데이터를 저장하는 함수까지
value1 = input("Address 입력하시오: ") # string


cond1 = (df1["VIP회원"] == "VIP")
cond2 = (df1["Address"] == value1)

df1.loc[cond1 & cond2].to_excel("vip_member3.xlsx")
```

```python
#연령범위
value1 = int(input("연령 범위 입력(최소값): "))
value2 = int(input("연령 범위 입력(최대값): "))

cond1 = (df1["Age"] >= value1) & (df1["Age"] <= value2)
cond2 = (df1["VIP회원"] == "VIP")

df1.loc[cond1 & cond2].to_excel("vip_age_member.xlsx")
```

```python
# input, to_excel 함수를 활용하면
# 우리가 원하는 데이터를 추출하고 파일로 저장하는 로직을 쉽게 짤 수 있다.
```

```python
#연령범위
# 저장할 때는 총 구매 금액이 높은 고객 순으로 정렬
value1 = int(input("연령 범위 입력(최소값): "))
value2 = int(input("연령 범위 입력(최대값): "))

cond1 = (df1["Age"] >= value1) & (df1["Age"] <= value2)
cond2 = (df1["VIP회원"] == "VIP")

df1.loc[cond1 & cond2].sort_values(by="총구매금액", ascending=False).to_excel("vip_age_member.xlsx")
```

```python


# 총 구매 수량이 100건 이상인 고객 중에 
# 총 구매 금액이 높은 고객 데이터
```

```python
# 총 구매 수량이 100건 이상이면서
# 총 구매 금액 상위 50명 고객 데이터
# 데이터파일 저장 ->  송부 (메일로 보내다)
cond1 = (df1["총구매수량"] >= 100)
df2 = df1.loc[cond1].sort_values(by="총구매금액", ascending=False).head(50)
df2.to_excel("top_sales_members.xlsx")
```
