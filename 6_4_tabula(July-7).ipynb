```python
# list가 아니라 df로 불러오기: 0번 째 table만 가져와보자
# 13페이지에는 테이블이 한개 있다  
# tabula.read_pdf(multiple_tables): 한 페이지에 표가 여러 개 있을 때, 여러 개의 표를 구분해서 가져오는 옵션
# [0]: 구분한 여러 테이블 중 0번 째 테이블
df1 = tabula.read_pdf("dataset/2012년_출판시장통계.pdf", pages=13, multiple_tables=True)[0]
```

```python
df1 # 컬럼 이름이 난잡하게 되어 있다. 
# 왜 난잡하게 잡혔냐? 
# pdf 상 테이블을 보면, 표가 연도별로 데이터가 나뉘어져 있기 때문이다.
# unnamed: 0, ...를 rename() 사용해서 컬럼명 바꿔준다.
```

```python
# 이름 바꿀 col이 많으면 rename 사용하기 어렵다. 시간적 소모. ㅁ
df1.rename(columns={"Unnamed: 0": "영업이익순위"})
# 한번에 바꾸는 방법이 있다. 
```

```python
col_names = df1.columns.tolist()
new_col_names = ['순위', '기업명', '분류', '매출액_2011', "영업이익_2011", "영업이익율_2011",
                "매출액_2012", "영업이익_2012", "영업이익율", "영업이익증감율"]
```

```python
# zip()
zip(col_names, new_col_names)
```

```python
dict(zip(col_names, new_col_names)) # 매칭되서 하나의 dict가 나오게 된다. 
```

```python
trans_name = dict(zip(col_names, new_col_names))
```

```python
# 캬.. 
# 0-3 row는 필요 없다. 
df1.rename(columns=trans_name)
```

```python
df1.rename(columns=trans_name).iloc[4:]
# 중간 중간에 소계1 같은 행이 있다. 이런 행은 제거해주자.
```

```python
# 앞에서 배운 함수들이 조금 조금씩 쓰인다. 
# 그렇게 더러운 데이터가 깔금하게 바뀌었다. 

# 지금은 선생님이 가이드를 해줬지만
# 실무에서는 실무 데이터를 갖고 여러분이 직접 연습(배운 함수들 쓰는거)을 해봐야 한다. 
df1.rename(columns=trans_name).iloc[4:].dropna()
# 문제1. 매출액이 겹친다. 
# 문제2. 매출액 중간에 comma가 있고 object 타입인 col이 많다.
# 문제3. 매출액_2011, 매출액_2012와 같이 매출액 col 하나로 묶을 수 있는 column이 있다. (연도는 다르지만 항목값이 중복되어있다.)
```

```python
df2 = df1.rename(columns=trans_name).iloc[4:].dropna()
```

```python
df2.columns
```

```python
# 숫자로 바꿔보자. comma 빼고 숫자로 변환
# comma 때문에 Python이 아래 ser를 object 타입으로 인식하고 있는 것이다.
df2["매출액_2011"]
```

```python
# apply '정규식'
# replace : 특정 문자를 제거하는 함수
def func1(row):
    return int(row.replace(',', ''))
```

```python
df2[""] = df2["매출액_2011"].apply(func1)
```

```python
# 연도, 매출액을 분리해보자. 
# 매출액_2011, 매출액_2012 -> 매출액, 연도
```

```python
# melt: 데이터를 한줄로 모아주는 함수
# pd.melt(id_vars): 변경하지 않을 col들, 그대로 냅둘 col 이름을 적는다. 
pd.melt(df2, id_vars=["순위", "기업명", "분류", "영업이익율", "영업이익증감율"])
# id_vars에 적은 col 제외하고, 나머지 col은 뒤에 연도가 붙어있다. 
# 연도가 붙은 col을 varivale 컬럼 하나에 몰아버리고, 그에 관련된 숫자 데이터를 value에 넣는다. 
```

```python
# variable 열에서 '_' 기준으로 '구분자'={'영업이익', '매출액'}, '연도'={2011, 2012} 데이터, 열 2개를 만든다.
df3 = pd.melt(df2, id_vars=["순위", "기업명", "분류", "영업이익율", "영업이익증감율"])
```

```python
df3["variable"]
```

```python
# split: 특정 문자를 기준으로 문자열을 list로 나누는 함수
# 특정 문자 기준으로 좌측 문자열은 lst[0]에, 우측 문자열은 lst[1]에 들어간다.
def func2(row):
    return row.split('_')
```

```python
df3["variable"].apply(func2)
```

```python
#lst[1]에 연도 데이터가 있으므로, idx=1로 연도 col을 만든다.
def func2(row):
    return row.split('_')[1]
```

```python
df3["연도"] = df3["variable"].apply(func2)
```

```python
df3
```

```python
def func3(row):
    return row.split('_')[0]
```

```python
df3["구분"] = df3["variable"].apply(func3)
```

```python
# melt: id_vars에 적지 않은 col 이름을 variable열 하나에 뭉침
# apply: split('_') 해서 (매출액, 영업이익)을 구분하고, (2011, 2012)를 구분함.
df3
```

```python
df3["구분"]
```

```python
cond1 = (df3["구분"] == "매출액")
df3.loc[cond1] # 이 df의 value 값은 모두 매출액이다. (2011 또는 2012년도)
```

```python
cond1 = (df3["구분"] == "매출액")
df3_1 = df3.loc[cond1].rename(columns={"value":"매출액"})
```

```python
df3_1
```

```python
cond1 = (df3["구분"] == "영업이익")
df3_2 = df3.loc[cond1].rename(columns={'value':"영업이익"})
```

```python
df3_2
```
